```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# a.
# arange(0,n) defaults to natural number steps ie. [0,1,2,3,...,n-1] NOT linspace()
n = np.arange(0, 100)
print(n)
y = 0.1 * n ** 2 - 10 * n + 5
plt.plot(n, y)
plt.show()
```

```python
# b.
y[0] = 1
y[1] = 1
for n in range(2, 100):
    y[n] = y[n - 1] + 0.5 * y[n - 2]
plt.plot(np.arange(0,100), y)
plt.show()
```

```python
# c.
y = np.zeros(100)
y[0] = 1
for n in range(1, 100):
    for m in range(n):
        y[n] = y[n] + (y[m] / (n - m))
    print("n:", n, "y(n):", y[n])
plt.plot(np.arange(0,100), y)
plt.show()
```

```python
# d.
y[0] = 3
x = np.arange(0,100)
for n in range(1, 100):
    if n % 3 == 0:
        y[n] = y[n - 1] / 3.0
    else:
        y[n] = 2 * y[n - 1] - 1
    print("x : ", x, "y: ", y)

plt.plot(x, y)
plt.plot()
```
