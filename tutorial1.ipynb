```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(1,10)
x1
```

```python
x2 = np.arange(1,10,0.1)
x2
```

```python
x3 = np.linspace(1,10,100)
x3
```

```python
y = x2*2
y
```

```python
y1 = x1**2-4*x1+2
plt.plot(x1,y1)
plt.plot(x2,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = 2x")
```

```python
x4 = np.linspace(1,10,100)
y = x4**3+3*x4-1
plt.plot(x4,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^3+3x-1")
```
