```python
# Import required libraries for plotting
import matplotlib.pyplot as plt
import numpy as np

# Define the function and its derivative using lambda functions for simplicity
# f represents the equation we want to find the root of (x^3 - x - 1 = 0)
f = lambda x: x**3 - x - 1

# def f(x):
#   return x**3 - x - 1

# f_prime is the derivative of f, used to adjust our guesses in the Newton-Raphson method
f_prime = lambda x: 3 * x**2 - 1

# def f_prime(x):
#   return 3 * x**2 - 1

# Starting point for our guess of the root, chosen as 1.5
x0 = 1.5

# Iteration parameters
# tolerance defines how close we need to get to the actual root before stopping
tolerance = 1e-20

# max_iterations limits the number of times we'll try to find the root
max_iterations = 15

# List to store details of each iteration (for demonstration and later display)
iterations = []

# Newton-Raphson iteration loop
for i in range(max_iterations):
    # Calculate the next guess using the Newton-Raphson formula:
    # x1 = x0 - f(x0) / f_prime(x0)
    x1 = x0 - f(x0) / f_prime(x0)

    # Store the current iteration details: iteration number, current guess, next guess,
    # function value at the current guess, and derivative value at the current guess
    iterations.append((i + 1, x0, x1, f(x0), f_prime(x0)))

    # Check if the change between our new guess and the old one is small enough (within tolerance)
    if abs(x1 - x0) < tolerance:
        # If so, we're close enough to the root, so break out of the loop
        break

    # Update x0 to be the new guess for the next iteration
    x0 = x1

# Display the results without using pandas, printing each iteration in a readable format
print("Iteration\t x_n\t\t x_(n+1)\t\t f(x_n)\t\t f'(x_n)")
for iteration in iterations:
    print(f"{iteration[0]}\t\t {iteration[1]:.6f}\t {iteration[2]:.6f}\t {iteration[3]:.6f}\t {iteration[4]:.6f}")

# Generate values for x to plot the function
x_values = np.linspace(-2, 2, 400)
y_values = f(x_values)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the function f(x)
plt.plot(x_values, y_values, label=r"$f(x) = x^3 - x - 1$", color="blue")

# Highlight the x-axis for reference
plt.axhline(0, color='black',linewidth=0.5)

# Plot each Newton-Raphson iteration as a point
for iteration in iterations:
    x_n = iteration[1]
    plt.plot(x_n, f(x_n), 'ro')  # Red point for each guess

# # Draw arrows to indicate the steps from one guess to the next
# for i in range(len(iterations) - 1):
#     x_n = iterations[i][1]
#     x_next = iterations[i][2]
#     plt.annotate('', xy=(x_next, 0), xytext=(x_n, f(x_n)),
#                  arrowprops=dict(arrowstyle='->', color='red'))

# Add labels and a legend
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Newton-Raphson Iterations for f(x) = x^3 - x - 1")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
