```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# create a new image
rows, cols = 360, 360
img1 = np.zeros((rows, cols))
```

```python
# place some squares
img1[10:30,20:40] = 0.5
img1[200:250,120:170] = 0.8
img1[82:104,60:82] = 1
img1[30:55,320:345] =0.3

plt.imshow(img1)
plt.colorbar()
plt.show()
```

```python
plt.imshow(img1,cmap="gray")
plt.colorbar()
plt.show()
```

```python
# crop the image
img2 = img1[0:240,0:240]
plt.imshow(img2)
plt.colorbar()
plt.show()
```

```python
img3 = np.zeros((rows, cols))
for m in range(0, rows):
    for n in range (0, cols):
        img3[m,n] = img1[rows - m - 1, cols - n - 1]
plt.imshow(img3)
plt.colorbar()
plt.show()
```

```python
# rotate image by 90 deg.
# see: https://en.wikipedia.org/wiki/Rotation_matrix
rotMtrx = np.zeros((2, 2))
angle = np.pi/2
rotMtrx[0, 0] = np.cos(angle)
rotMtrx[0, 1] = np.sin(angle)
rotMtrx[1, 0] = -np.sin(angle)
rotMtrx[1, 1] = np.cos(angle)
rotMtrx
```

```python
#a = [10,30]
#print(np.matmul(rotMtrx,a))

a = np.transpose([0,0])
img2 = np.zeros((360,360))
for m in range(0, rows):
    for n in range (0, cols):
        a = np.matmul(rotMtrx, [n, m])
        # print(a)
        img2[int(a[0]),int(a[1])] = img1[n, m]
plt.imshow(img2)
plt.colorbar()
plt.show()
```

```python
img3 = np.zeros((360,360))
r = 180
c = 180
for m in range(0, rows):
    for n in range (0, cols):
        if np.sqrt((m - c) ** 2 + (n - c) ** 2)<r:
            img3[m,n] = img1[m,n]
plt.imshow(img3)
plt.colorbar()
plt.show()
```
